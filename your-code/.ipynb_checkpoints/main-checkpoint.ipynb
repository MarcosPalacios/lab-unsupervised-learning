## Unskewing data

```python
NumCols = df.columns[2:]
```

```python
df[NumCols].skew()
```

```python
#Usamos logaritmo para corregir la skewness.

df_log = np.log(df[NumCols])
df_log.skew()
```

```python
#Usamos la raiz cuadrada para corregir la skewness.

df_sqrt = np.sqrt(df[NumCols])
df_sqrt.skew()
```

```python
#Usamos boxcox para corregir la skewness ya que todos >0
from scipy import stats

df_boxcox1 = pd.DataFrame([pd.Series(stats.boxcox(df[col])[0],name=col) for col in NumCols]).T

df_boxcox1.skew()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data = df_boxcox1)
```

```python
#Usamos boxcox con otra libreria para corregir la skewness ya que todos >0
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='box-cox')
#Este metodo ademas de aplicar boxcox estandariza los valores.

df_boxcox2 = pd.DataFrame(pt.fit_transform(df[NumCols]),columns = NumCols)

df_boxcox2.skew()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data = df_boxcox2)
```

```python
#BoxCox es la que mejor resultados nos da utilizamos esa.
```

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:
from sklearn.preprocessing import StandardScaler
# Your code here:

#Necesitamos estandarizar para igualar medias y varianzas por requisito de kmeans.

#No tiene  sentido Estandarizar las variables dummies, solo se estandarizan las variables numericas
customers_scale = pd.DataFrame(StandardScaler().fit_transform(df_boxcox1),columns = NumCols) 
```

# Challenge 3 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

# Inertia is the sum of squared error for each cluster. 
# Therefore the smaller the inertia the denser the cluster(closer together all the points are)

inertias = [KMeans(k).fit(df_std).inertia_ for k in range(2,10)]

plt.figure(figsize=(8,8))
plt.plot(range(2,10), inertias)

#No hay un codo claro, seguramente debido a que no hay clusters en nuestros datos.
```

```python
# Your code here:
from sklearn.cluster import KMeans
k = 5 
cl = KMeans(k)
cl.fit(df_std)
```

Count the values in `labels`.

```python
clusters = cl.predict(df_std) #Equivalente a cl.labels_
```

```python
# Your code here:

df['labels'] = cl.labels_

df['labels'].value_counts()
```

```python
#No hay clusters claros.
sns.pairplot(data = df ,hue='labels')
```

```python
#No hay clusters claros.
sns.pairplot(data = df[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen','labels']] ,hue='labels')
```

# Challenge 4 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
# Your code here
from sklearn.cluster import DBSCAN

dbs = DBSCAN(eps=0.5, min_samples=5)

dbs.fit(df_std)
```

```python
dbs.labels_ #Raro, retorna todos 1
```

Count the values in `labels_DBSCAN`.

```python
# Your code here:
df['labels_DBSCAN'] = dbs.labels_

df['labels_DBSCAN'].value_counts()
```
